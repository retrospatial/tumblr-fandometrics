# Fandoms Analysis

```python
import pandas as pd

df = pd.read_csv('ships_complete.csv', encoding='latin1')
df.head()
```

## highrankers (#1 ship from each year)

```python
# groupby creates two groups, where rank 1 is either True or False
# get_group only retrieves the True group, then we obtain the names

rank_1 = df.groupby(df['rank']==1).get_group(True)['name']

# turn it into a list
rank_1_list = [name for name in rank_1]
rank_1_list
```

```python
# create a dataframe named rank_1_df and populate it with just the rank 1 ships
# use set() to keep unique names and pd.concat to keep adding to the dataframe
# reset the index

rank_1_df = pd.DataFrame()

for name in set(rank_1_list):
    rank_1_df = pd.concat([rank_1_df, df[df['name'] == name]])

rank_1_df = rank_1_df.reset_index(drop=True)

rank_1_df
```

```python
rank_1_df.to_csv('rank_1.csv', index=False, encoding='latin1')
```
