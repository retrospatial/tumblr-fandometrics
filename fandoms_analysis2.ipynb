# canon is real (?)

```python
import pandas as pd

df = pd.read_csv('ships_complete.csv', encoding='latin1')
df.head()
```

```python
len(df['name'].unique())
```

```python
is_canon = df[df['canon'] == 'Yes']
is_canon_names = set(is_canon['name'])
canon_length= len(is_canon_names)
canon_percentage = round((canon_length / (38 + 58)) * 100)

print(f"Canon ships make up {canon_percentage}% of ships")
```

```python
not_canon = df[df['canon'] == 'No']
not_canon_names = set(not_canon['name'])
not_canon_length = len(not_canon_names)
not_canon_percentage = round((not_canon_length / (38 + 58)) * 100)

print(f"Canon ships make up {not_canon_percentage}% of ships")
```

# M/M canon vs non canon

```python
slash_canon = df[(df['pairs'] == 'M/M') & (df['canon'] == 'Yes')]
slash_canon_unique = slash_canon['name'].unique()
slash_canon_length = len(slash_canon_unique)

slash_not_canon = df[(df['pairs'] == 'M/M') & (df['canon'] == 'No')]
slash_not_canon_unique = slash_not_canon['name'].unique()
slash_not_canon_length = len(slash_not_canon_unique)

total = slash_canon_length + slash_not_canon_length

print(f"Canon ships make up {round(slash_canon_length/total*100)}% of M/M pairings")
print(f"Non canon ships make up {round(slash_not_canon_length/total*100)}% of M/M pairings")
```

```python
print("Slash canon:", slash_canon_length)
print("Slash not canon:", slash_not_canon_length)
```

```python
femslash_canon = df[(df['pairs'] == 'F/F') & (df['canon'] == 'Yes')]
femslash_canon_unique = femslash_canon['name'].unique()
femslash_canon_length = len(femslash_canon_unique)

femslash_not_canon = df[(df['pairs'] == 'F/F') & (df['canon'] == 'No')]
femslash_not_canon_unique = femslash_not_canon['name'].unique()
femslash_not_canon_length = len(femslash_not_canon_unique)

total = femslash_canon_length + femslash_not_canon_length

print(f"Canon ships make up {round(femslash_canon_length/total*100)}% of F/F pairings")
print(f"Non canon ships make up {round(femslash_not_canon_length/total*100)}% of F/F pairings")
```

```python
print("Femslash canon:", femslash_canon_length)
print("Femslash not canon:", femslash_not_canon_length)
```
